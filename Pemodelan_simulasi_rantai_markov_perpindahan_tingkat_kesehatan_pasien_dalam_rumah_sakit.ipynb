```python
import numpy as np
import matplotlib.pyplot as plt

# Parameter awal
P_krit_0 = 50        # Pasien kritis awal
P_stabil_0 = 30      # Pasien stabil awal
P_sembuh_0 = 10      # Pasien sembuh awal
r_krit_stabil = 0.05 # Laju perpindahan dari kritis ke stabil
r_stabil_sembuh = 0.1# Laju perpindahan dari stabil ke sembuh
t = np.linspace(0, 50, 500) # Waktu (hari)

# Fungsi perpindahan
P_krit = P_krit_0 * np.exp(-r_krit_stabil * t)
P_stabil = (P_stabil_0 + (P_krit_0 * r_krit_stabil / (r_stabil_sembuh - r_krit_stabil)) *
           (np.exp(-r_krit_stabil * t) - np.exp(-r_stabil_sembuh * t)))
P_sembuh = P_sembuh_0 + P_krit_0 - P_krit - P_stabil

# Plot
plt.figure(figsize=(12, 8))
plt.plot(t, P_krit, label='Pasien Kritis', color='red')
plt.plot(t, P_stabil, label='Pasien Stabil', color='orange')
plt.plot(t, P_sembuh, label='Pasien Sembuh', color='green')
plt.title('Perpindahan Tingkat Kesehatan Pasien di Rumah Sakit')
plt.xlabel('Waktu (hari)')
plt.ylabel('Jumlah Pasien')
plt.legend()
plt.grid()
plt.show()
```
